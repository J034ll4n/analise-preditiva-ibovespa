```python
import pandas as pd
import os

# Lista dos nomes dos arquivos que você quer juntar
nomes_dos_arquivos = [
    'data/08-2000-08-1970.csv',
    'data/08-2015-08-2000.csv',
    'data/08-2015-08-2025.csv'
]

# Lista para armazenar os dataframes de cada arquivo
lista_de_dataframes = []

# Loop para ler cada arquivo e adicionar o dataframe na lista
for nome_arquivo in nomes_dos_arquivos:
    # Verifica se o arquivo existe no diretório
    if os.path.exists(nome_arquivo):
        df = pd.read_csv(nome_arquivo)
        lista_de_dataframes.append(df)
    else:
        print(f"Aviso: O arquivo '{nome_arquivo}' não foi encontrado e será ignorado.")

# Verifica se algum arquivo foi lido antes de tentar juntar
if lista_de_dataframes:
    # Junta todos os dataframes da lista em um único dataframe
    dataframe_final = pd.concat(lista_de_dataframes, ignore_index=True)

    # Salva o dataframe final em um novo arquivo CSV
    dataframe_final.to_csv('Dados_Historicos_Ibovespa_Consolidado.csv', index=False)

    print("Os arquivos foram juntados com sucesso!")
    print("O resultado foi salvo no arquivo 'Dados_Historicos_Ibovespa_Consolidado.csv'")
else:
    print("Nenhum arquivo de entrada foi encontrado. A operação foi cancelada.")
```
